```python
import csv
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize': (18, 6)})
```

```python
import etl as l
```

```python


dat = l.df5
dat = dat[dat['Year'] == 2019]
dat = dat[(dat['BrandGroup'] == 'T1') | (dat['BrandGroup'] == 'VIZIO') | (dat['BrandGroup'] == 'TCL')]

dat['YearMonth'] = dat['Year'].astype(str) + dat['Month'].astype(str)

dat = dat[dat['ASP'] > 0]

dat.head()
```

```python
dat.info()
```

```python
sns.set(style="darkgrid")

# data = data[data['BrandGroup'] != 'Other']
# df4 = df4[df4['BrandGroup'] != 'OPP']

sns.lineplot(x="YearMonth", y="ASP",
             hue="BrandGroup",
             data=dat).set_title('ASP Range')
```

```python
sns.lineplot(x="YearMonth", y="ASP",
             hue="Size (group)",
             data=dat).set_title('ASP Range')
```

```python
temp = dat[(dat['Size (group)'] != '76+') & (dat['Size (group)'] != '75') & (dat['Size (group)'] != 'Other')]

sns.lineplot(x="YearMonth", y="ASP",
             hue="Size (group)",
             data=temp).set_title('ASP Range')
```

```python
temp = dat[(dat['Size (group)'] != '76+') & (dat['Size (group)'] != '75') & (dat['Size (group)'] != 'Other')]

sns.lineplot(x="YearMonth", y="ASP",
             hue="Size (group)",
             data=temp).set_title('ASP Range')
```

```python
s_32 = dat[dat['Size (group)'] == 32]
s_32 = s_32[s_32['Year'] == 2019]
sns.lineplot(x="YearMonth", y="ASP",
             hue="Brand",
             data=s_32).set_title('ASP Range')
```

```python
s_50 = dat[dat['Size (group)'] == '46 - 52']
# s_50 = s_32[s_32['Year'] == 2019]
sns.lineplot(x="YearMonth", y="ASP",
             hue="Brand",
             data=s_50).set_title('ASP Range 50"')
```

```python
s_55 = dat[dat['Size (group)'] == '54 - 58']
s_55 = s_55[s_55['Sub Res'] == 'UHD']
sns.lineplot(x="YearMonth", y="ASP",
             hue="Brand",
             data=s_55).set_title('ASP Range 55"')
```

```python
s_55 = dat[dat['Size (group)'] == '54 - 58']
s_55 = s_55[s_55['Sub Res'] == 'UHD']

s_55['ASP_rolling'] = s_55['ASP'].rolling(3).mean()
sns.lineplot(x="YearMonth", y="ASP_rolling",
             hue="Brand",
             data=s_55).set_title('ASP Range Rolling 3 Mos 55"')
```

```python
s_65 = dat[dat['Size (group)'] == '65']
s_65 = s_65[s_65['Sub Res'] == 'UHD']

s_65['ASP_rolling'] = s_65['ASP'].rolling(3).mean()
sns.lineplot(x="YearMonth", y="ASP_rolling",
             hue="Brand",
             data=s_65).set_title('ASP Range Rolling 3 Mos 65"')

# s_65['ASP_rolling']['Brand' == 'SAMSUNG'].describe()
```
